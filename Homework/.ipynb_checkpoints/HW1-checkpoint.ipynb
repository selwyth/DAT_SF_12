####3. Use Pandas "Read JSON" functionality to turn the 311 data into a Pandas Dataframe and display the first 20 rows


```python
#YOUR CODE HERE
data_pd = pd.read_json(urllib.urlopen(url))
pd.set_option('display.max_rows', 20)
data_pd.head(n = 20)
```

####4. How many missing values do we have? 

```python
#YOUR CODE HERE
print data_pd.isnull().sum() # this returns the number of missing values by column
print """ 
      I think you want the missing values by column, 
      not total missing values, 
      but just in case you want the latter, 
      the total is %i.
      """ % data_pd.isnull().sum().sum()
```

####5. Drop the "point" feature from the dataframe - its a dictionary, which is unsupported in a flat database

```python
#YOUR CODE HERE
print len(data_pd.columns), data_pd.columns #DF 15 features to start
data_pd2 = data_pd.drop('point', axis = 1)
print len(data_pd2.columns), data_pd2.columns #DF 14 features remaining, with point gone; looks good
```

####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
data_pd3 = data_pd2.dropna(how = 'all') #DF get rid of rows with missing values
data_pd3 = data_pd3.drop(['media_url'], axis = 1) 
#DF I'm forced to drop this column because it has dicts, and I get errors when loading into SQL database
```

```python
#YOUR CODE HERE
ram_con = sqlite3.connect(':memory:')
data_pd3.to_sql('data_db', ram_con, if_exists = 'replace')
```

###### 6. Display the first 5 rows of the DB

```python
#YOUR CODE HERE
pd.read_sql("SELECT * FROM data_db LIMIT 5", ram_con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#YOUR CODE HERE
pd.read_sql("SELECT * FROM data_db WHERE neighborhood = 'Haight Ashbury' AND category like '%Sidewalk Cleaning%' LIMIT 10", ram_con)
# pd.read_sql("SELECT distinct category, request_type, request_details FROM data_db WHERE neighborhood = 'Haight Ashbury'", ram_con)
```

####8. Which type of complaint was most common? 

```python
#YOUR CODE HERE
pd.read_sql("""
    SELECT category, count(*)
    FROM data_db
    GROUP BY category
    ORDER BY count(*) DESC
    LIMIT 1
    """, ram_con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE
pd.read_sql("""
    SELECT    neighborhood
            , count(*) as total_complaints
            , sum(case when category like '%Sidewalk Cleaning%' then 1 else 0 end) as sidewalk_cleaning
            , cast(
              sum(case when category like '%Sidewalk Cleaning%' then 1 else 0 end) as float) / count(*) * 100 as perc
    FROM data_db
    GROUP BY neighborhood
    ORDER BY sidewalk_cleaning ASC
    """, ram_con
    )
```

```python
#YOUR CODE HERE
pd.read_sql("""
    SELECT    neighborhood
            , count(*) as total_complaints
            , sum(case when category like '%Sidewalk Cleaning%' then 1 else 0 end) as sidewalk_cleaning
            , cast(
              sum(case when category like '%Sidewalk Cleaning%' then 1 else 0 end) as float) / count(*) * 100 as perc
    FROM data_db
    GROUP BY neighborhood
    ORDER BY perc ASC
    """, ram_con
    )
```
