# kNN

```python
# reading wine.csv into memory
import pandas as pd
wine = pd.read_csv('../data/wine.csv')
print wine.head()
wine.describe()
```

```python
wine.Wine.value_counts() # looks like 3 wine classes
```

```python
# splitting data into train and 30% test with random_state 42
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
                                                    wine.iloc[ :, 1:len(wine.columns) ],
                                                    wine.Wine,
                                                    test_size=0.3,
                                                    random_state=42)

print X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
# apply kNN
from sklearn import neighbors, feature_selection
from seaborn import plt
%matplotlib inline

n_neighbors = range(1,80)

scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
    
plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
```

1, 27, 29 etc provide highest score. Let's pick 1 nearest neighbor. I did try 27 neighbors too, and it provides the same accuracy but different rows are predicted differently, so need further examination to determine how many nearest neighbors is better if I dig deeper.

```python
clf_3 = neighbors.KNeighborsClassifier(n_neighbors = 1)
clf_3.fit(X_train, y_train)
score = clf_3.score(X_test, y_test)
print "My accuracy is {:.2f} using 1 nearest neighbors.".format(score)
```

```python
comparison = pd.DataFrame({'true value': y_test ,
                           'prediction' : clf_3.predict(X_test) })
features = pd.DataFrame(X_test, columns = wine.columns[1:len(wine.columns)])

comparison = pd.concat([comparison, features], axis = 1)

print comparison.iloc[:, 0:2]
comparison.head()
```

# k-Means

```python
from sklearn.cluster import KMeans

mod = KMeans(n_clusters = 3, random_state = 42)

# I'm going to fit the entire data set this time instead of just the training set since this is an unsupervised problem

mod.fit(wine.iloc[ :, 1:len(wine.columns) ])

y_hat_kmeans = mod.labels_

comparison_kmeans = pd.DataFrame({'actual category': wine.iloc[:, 0] , 'cluster': y_hat_kmeans})

print comparison_kmeans
```

```python
# checking which of my clusters corresponds to which category
comparison_kmeans['count'] = 1
comparison_kmeans.groupby(['actual category', 'cluster']).count()
```

looks like category 1 corresponds to my cluster 1, category 2 corresponds to my cluster 0, category 3 corresponds to my cluster 2

```python
def check_accuracy(actual, pred):
    correct = []
    for i, j in zip(actual, pred):
        if i == 1 and j == 1:
            correct.append(1)
        elif i == 2 and j == 0:
            correct.append(1)
        elif i == 3 and j == 2:
            correct.append(1)
        else:
            correct.append(0)
            
    return correct
            
accu = check_accuracy(wine.iloc[:,0], y_hat_kmeans)

print 'k-means (3) predicted the right cluster for {0} out of {1} cases, or {2:.1f}%.'.format(sum(accu), len(accu), float(sum(accu))/len(accu) * 100)
```

# Hierarchical Clustering

```python
from scipy.spatial.distance import pdist, squareform

distx = squareform(pdist(wine.iloc[:, 1:len(wine.columns)].values, metric='euclidean'))
distx
```

```python
# single method

from scipy.cluster.hierarchy import linkage, dendrogram

R = dendrogram(linkage(distx, method='single'))

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
# ward method

from scipy.cluster.hierarchy import linkage, dendrogram

R = dendrogram(linkage(distx, method='ward'), color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```
